```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image and the filter/template
image = cv2.imread('mountain.jpg', cv2.IMREAD_GRAYSCALE)
filter_template = cv2.imread('mountain.jpg', cv2.IMREAD_GRAYSCALE)

# Perform correlation using OpenCV's matchTemplate function

correlation_map = cv2.matchTemplate(image, filter_template, cv2.TM_CCOEFF_NORMED)

# Find the location of the best match
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(correlation_map)
top_left = max_loc
bottom_right = (top_left[0] + filter_template.shape[1], top_left[1] + filter_template.shape[0])

# Draw a rectangle around the best match in the original image
result_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.rectangle(result_image, top_left, bottom_right, (0, 255, 0), 2)

# Display the result
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(result_image)
plt.title('Correlation Result')
plt.axis('off')

plt.tight_layout()
plt.show()
```
